```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
e1=le.fit_transform(df2.workclass)
e2=le.fit_transform(df2.education)
e3=le.fit_transform(df2["marital.status"])
e4=le.fit_transform(df2.occupation)
e5=le.fit_transform(df2.relationship)
e6=le.fit_transform(df2.race)
e7=le.fit_transform(df2.sex)
e8=le.fit_transform(df2["native.country"])
e9=le.fit_transform(df2.income)
e8
```

```python
features = list(zip(e1,e2,e3,e4,e5,e6,e7,df2["hours.per.week"],e8,df2["age.group"]))
print(features)
```

```python
label = e9
label
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.3,random_state=12000)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
gnb_acc=metrics.accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,confusion_matrix,f1_score
print("precision score : "+str(precision_score(y_test, y_pred))) # tp/tp+fp
print("accuracy score : "+str(accuracy_score(y_test, y_pred))) # total correct 
print("recall score : "+str(recall_score(y_test, y_pred)))   # tp/tp+fn
print("f1 score : "+str(f1_score(y_test, y_pred))) 
average_precision=precision_score(y_test, y_pred)
```

```python
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt



precision, recall, _ = precision_recall_curve(y_test, y_pred)

plt.step(recall, precision, color='b', alpha=0.2,where='post')
plt.fill_between(recall, precision, step='post', alpha=0.2,color='b')

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('2-class Precision-Recall curve: AP={0:0.2f}'.format(average_precision))
```
